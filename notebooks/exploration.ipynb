Creating a simple Graph

```python
import pandas as pd
import numpy as np

train = pd.read_csv("../data/raw/train.csv")

train.shape
train.head()
train.info()
train.describe()
```

```python
target = "SalePrice"

num_cols = train.select_dtypes(include=["int64", "float64"]).columns
cat_cols = train.select_dtypes(include=["object"]).columns

len(num_cols), len(cat_cols)
missing = train.isnull().sum().sort_values(ascending=False)
missing[missing>0]

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(train["SalePrice"], kde=True)
plt.title("Histogram of SalePrice Frequency")
plt.xlabel("SalePrice")
plt.ylabel("Frequency")
plt.show()
plt.savefig('saleprice.png')

```

```python
log_price=np.log10(train["SalePrice"])
sns.histplot(log_price, kde=True)
plt.xlabel("log(SalePrice)")
plt.ylabel("Frequency")
plt.title("Transformed Graph")
plt.show()
```

```python

sns.boxplot(x=train["SalePrice"], color='skyblue')
plt.show()
```

```python
corr = train[num_cols].corr()["SalePrice"].sort_values(ascending=False)
corr.head(10)
```

```python
top_features = corr.head(10).index
sns.heatmap(train[top_features].corr(), annot=True, cmap="coolwarm")
plt.show()
'''least_features= corr.tail(10).index
sns.heatmap(train[least_features].corr(), annot=True, cmap="coolwarm")
plt.show()'''
```

```python


# 1. Scatter Plot for Overall Quality (Categorical/Ordinal)
sns.boxplot(x='OverallQual', y='SalePrice', data=train)
plt.title('Overall Quality vs SalePrice')
plt.show()

# 2. Scatter Plot for Living Area (Continuous)
sns.scatterplot(x='GrLivArea', y='SalePrice', data=train)
plt.title('Area(GrLivArea) vs SalePrice')
plt.show()
```

```python
for col in ["Neighborhood", "BldgType", "HouseStyle", "MSZoning"]:
    plt.figure(figsize=(10,4))
    sns.boxplot(x=col, y="SalePrice", data=train)
    plt.xticks(rotation=45)
    plt.show()
```
